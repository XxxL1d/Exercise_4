```python
import random
import time

# Define all cards
card_deck = [
    {"Name": "Diablo", "Health": 100, "Attack": 90, "Defense": 60},
    {"Name": "Medusa", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Jester", "Health": 120, "Attack": 60, "Defense": 90},
    {"Name": "Troll", "Health": 150, "Attack": 40, "Defense": 94},
    {"Name": "Specter", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Mist", "Health": 100, "Attack": 75, "Defense": 65},
    {"Name": "Savage", "Health": 100, "Attack": 90, "Defense": 50},
    {"Name": "Marauder", "Health": 100, "Attack": 85, "Defense": 50},
    {"Name": "Wimp", "Health": 110, "Attack": 40, "Defense": 85},
    {"Name": "Sorcerer", "Health": 100, "Attack": 70, "Defense": 55}
]

# Define the number of cards each player has
deck_size = 5

# Shuffle and distribute cards
random.shuffle(card_deck)
card_deck_copy = [card.copy() for card in card_deck]  # Deep copy to avoid referencing the same object

player_deck = card_deck_copy[:deck_size]
opponent_deck = card_deck_copy[deck_size:deck_size*2]

# Show player's cards
def show_player_deck():
    print("\nYour Cards:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}. {card['Name']} - Health: {card['Health']}, Attack: {card['Attack']}, Defense: {card['Defense']}")
    print()

# Handle attack logic
def process_attack(attacker, defender):
    attack_value = attacker['Attack']
    defense_value = defender['Defense']
    health_value = defender['Health']

    if attack_value > health_value and defense_value == 0:
        # Attack power exceeds health, instant defeat
        print(f"{attacker['Name']}'s attack power exceeds {defender['Name']}'s health!")
        defender['Health'] = 0
    else:
        if defense_value >= attack_value:
            # Defense is enough to block the attack
            defender['Defense'] -= attack_value
            print(f"{attacker['Name']} attacked {defender['Name']}, reducing its defense to {defender['Defense']}.")
        else:
            # Defense is insufficient, remaining attack reduces health
            damage_to_health = attack_value - defense_value
            defender['Defense'] = 0
            defender['Health'] -= damage_to_health
            print(f"{attacker['Name']} attacked {defender['Name']}, reducing defense to 0 and health to {defender['Health']}.")

# Game loop
round_num = 1
while len(player_deck) > 0 and len(opponent_deck) > 0:
    print(f"\n===== Round {round_num} =====")
    # Player's turn
    print("\n--- Your Turn ---")
    show_player_deck()
    opponent_card = random.choice(opponent_deck)
    print(f"The opponent played {opponent_card['Name']} - Health: {opponent_card['Health']}, Attack: {opponent_card['Attack']}, Defense: {opponent_card['Defense']}\n")

    # Player chooses a card
    while True:
        try:
            choice = int(input("Please choose your card (enter number): ")) - 1
            if 0 <= choice < len(player_deck):
                player_card = player_deck[choice]
                break
            else:
                print("Invalid choice, please enter again.")
        except ValueError:
            print("Please enter a valid number.")

    # Process player's attack
    if player_card['Attack'] > opponent_card['Health']:
        print(f"{player_card['Name']}'s attack power exceeds {opponent_card['Name']}'s health!")
        opponent_card['Health'] = 0
    else:
        process_attack(player_card, opponent_card)

    # Check if opponent's card is defeated
    if opponent_card['Health'] <= 0:
        print(f"{opponent_card['Name']} was defeated!")
        opponent_deck.remove(opponent_card)

    # Pause for 3 seconds
    time.sleep(3)

    # Check if the game is over
    if len(opponent_deck) == 0:
        print("Congratulations! You defeated all of the opponent's cards and won!")
        break

    # Opponent's turn
    print("\n--- Opponent's Turn ---")
    if len(opponent_deck) == 0:
        print("The opponent has no cards left! You win!")
        break

    opponent_card = random.choice(opponent_deck)
    print(f"The opponent played {opponent_card['Name']} - Health: {opponent_card['Health']}, Attack: {opponent_card['Attack']}, Defense: {opponent_card['Defense']}\n")

    show_player_deck()

    # Player chooses a card for defense
    while True:
        try:
            choice = int(input("Please choose a card to defend against the opponent's attack (enter number): ")) - 1
            if 0 <= choice < len(player_deck):
                player_card = player_deck[choice]
                break
            else:
                print("Invalid choice, please enter again.")
        except ValueError:
            print("Please enter a valid number.")

    # Opponent attacks player's card
    if opponent_card['Attack'] > player_card['Health']:
        print(f"{opponent_card['Name']}'s attack power exceeds {player_card['Name']}'s health!")
        player_card['Health'] = 0
    else:
        process_attack(opponent_card, player_card)

    # Check if player's card is defeated
    if player_card['Health'] <= 0:
        print(f"Your {player_card['Name']} was defeated!")
        player_deck.remove(player_card)

    # Pause for 3 seconds
    time.sleep(3)

    # Check if the game is over
    if len(player_deck) == 0:
        print("Unfortunately, all your cards were defeated. Game over.")
        break

    print("A new round is about to begin...\n")
    time.sleep(2)
    round_num += 1

# Game Over
if len(player_deck) > 0 and len(opponent_deck) == 0:
    print("You won the game!")
elif len(opponent_deck) > 0 and len(player_deck) == 0:
    print("The opponent won the game!")
else:
    print("Game over.")
    
```
